```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
```

```python
xl_file = pd.ExcelFile("Datasets/final_labels_SG2.xlsx")

df_load = pd.read_excel(xl_file, sheet_name='Sheet1')
```

```python
df_load
```

```python
df_load['label_bias'].value_counts()
```

```python
df_load.drop(df_load[df_load.label_bias == "No agreement"].index, inplace = True)
```

```python
df_load['label_bias'].value_counts()
```

```python
df = df_load.loc[:, ['text', 'label_bias']]
```

```python
df['label_bias'] = np.where(df['label_bias'] == "Biased", 1, 0)
df.head()
```

```python
df.rename(columns = {'label_bias':'bias'}, inplace = True)
```

```python
df.info()
```

```python
train_df, test_df = train_test_split(df, test_size=0.3, random_state=42)
train_df.shape, test_df.shape
```

```python
train_df
```

```python
train_df.to_csv('Datasets/Splits/Train/final_labels_SG2_label_bias_train.csv', index = False)
```

```python
test_df.to_csv('Datasets/Splits/Test/final_labels_SG2_label_bias_test.csv', index = False)
```
