```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
```

```python
from datasets import load_dataset

dataset = load_dataset("GonzaloA/fake_news")
```

```python
dataset
```

```python
df_train = dataset['train'].to_pandas()
df_train
```

```python
df = df_train
```

```python
df.drop(df.columns[df.columns.str.contains('unnamed',case = False)],axis = 1, inplace = True)
```

```python
df.head()
```

```python
df = df.drop(['text'], axis = 1)
```

```python
df.info()
```

```python
df.rename(columns={'title': 'text'}, inplace=True)
```

```python
df.rename(columns={'label': 'bluff'}, inplace=True)
```

```python
df['bluff'] = np.where(df['bluff'] == 0, 1, 0)
```

```python
df
```

```python
df.to_csv('Datasets/Fake_News_TFG.csv', index = False)
```

```python
train_df, test_df = train_test_split(df, test_size=0.30, random_state=42)
train_df.shape, test_df.shape
```

```python
train_df
```

```python
train_df.to_csv('Datasets/Splits/Train/FakeNewsTFG_label_lie_train.csv', index = False)
```

```python
test_df.to_csv('Datasets/Splits/Test/FakeNewsTFG_label_lie_test.csv', index = False)
```
