```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
```

```python
from datasets import load_dataset

dataset = load_dataset("fhamborg/news_sentiment_newsmtsc")
```

```python
dataset
```

```python
df_train = dataset['train'].to_pandas()
df_train
```

```python
df_val = dataset['validation'].to_pandas()
df_val
```

```python
df_test = dataset['test'].to_pandas()
df_test
```

```python
frames = [df_train, df_val, df_test]
df = pd.concat(frames, ignore_index = True)
df
```

```python
df = df.loc[:, ['sentence', 'polarity']]
```

```python
df.rename(columns={'sentence': 'text'}, inplace=True)
df.rename(columns={'polarity': 'sentiment'}, inplace=True)
```

```python
df.head()
```

```python
df = df.drop_duplicates(keep='first', ignore_index = True)
```

```python
df
```

```python
df = df.drop_duplicates(subset = ['text'], keep=False, ignore_index = True)
```

```python
df
```

```python
df['sentiment'].value_counts()
```

```python
def label (row):
   if row['sentiment'] == (-1):
      return 'negative'
   elif row['sentiment'] == 0:
      return 'neutral'
   elif row['sentiment'] == 1:
      return 'positive'
   else:
      return None
```

```python
df['sentiment'] = df.apply(lambda row: label(row), axis=1)
```

```python
df['sentiment'].value_counts()
```

```python
df['negative'] = ''
df['neutral'] = ''
df['positive'] = ''
```

```python
df.info()
```

```python
df.head()
```

```python
def label_sentiment (row):
   if row['sentiment'] == "negative":
      row['negative'] = 1
      row['neutral'] = 0
      row['positive'] = 0
      return 0
   elif row['sentiment'] == "neutral":
      row['negative'] =0
      row['neutral'] = 1
      row['positive'] = 0
      return 1
   elif row['sentiment'] == "positive":
      row['negative'] = 0
      row['neutral'] = 0
      row['positive'] = 1
      return 2
   else:
      return None
```

```python
df['sentiment'] = df.apply(lambda row: label_sentiment(row), axis=1)
```

```python
df
```

```python
df.to_csv('Datasets/NewsMTSC.csv', index = False)
```

```python
train_df, test_df = train_test_split(df, test_size=0.30, random_state=42)
train_df.shape, test_df.shape
```
