```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
```

```python
from datasets import load_dataset

dataset = load_dataset("liar")
```

```python
dataset
```

```python
df_train = dataset['train'].to_pandas()
df_train
```

```python
df = df_train
```

```python
df = df.loc[:, ['statement', 'label']]
```

```python
df.rename(columns={'statement': 'text'}, inplace=True)
```

```python
df.head()
```

```python
df['label'].value_counts()
```

```python
def label (row):
   if row['label'] == 0:
      return "false"
   elif row['label'] == 1:
      return "half-true"
   elif row['label'] == 2:
      return "mostly-true"
   elif row['label'] == 3:
      return "true"
   elif row['label'] == 4:
      return "barely-true"
   elif row['label'] == 5:
      return "pants-fire"
   else:
      return None
```

```python
df['label'] = df.apply(lambda row: label(row), axis=1)
```

```python
df['label'].value_counts()
```

```python
df['true'] = ''
df['mostly-true'] = ''
df['half-true'] = ''
df['barely-true'] = ''
df['false'] = ''
df['pants-fire'] = ''
```

```python
df.info()
```

```python
def label_lie (row):
   if row['label'] == "true":
      row['true'] = 1
      row['mostly-true'] = 0
      row['half-true'] = 0
      row['barely-true'] = 0
      row['false'] = 0
      row['pants-fire'] = 0
      return 0
   elif row['label'] == "mostly-true":
      row['true'] = 0
      row['mostly-true'] = 1
      row['half-true'] = 0
      row['barely-true'] = 0
      row['false'] = 0
      row['pants-fire'] = 0
      return 1
   elif row['label'] == "half-true":
      row['true'] = 0
      row['mostly-true'] = 0
      row['half-true'] = 1
      row['barely-true'] = 0
      row['false'] = 0
      row['pants-fire'] = 0
      return 2
   elif row['label'] == "barely-true":
      row['true'] = 0
      row['mostly-true'] = 0
      row['half-true'] = 0
      row['barely-true'] = 1
      row['false'] = 0
      row['pants-fire'] = 0
      return 3
   elif row['label'] == "false":
      row['true'] = 0
      row['mostly-true'] = 0
      row['half-true'] = 0
      row['barely-true'] = 0
      row['false'] = 1
      row['pants-fire'] = 0
      return 4
   elif row['label'] == "pants-fire":
      row['true'] = 0
      row['mostly-true'] = 0
      row['half-true'] = 0
      row['barely-true'] = 0
      row['false'] = 0
      row['pants-fire'] = 1
      return 5
   else:
      return None
```

```python
df['label'] = df.apply(lambda row: label_lie(row), axis=1)
```

```python
df
```

```python
df.to_csv('Datasets/LIAR.csv', index = False)
```

```python
train_df, test_df = train_test_split(df, test_size=0.30, random_state=42)
train_df.shape, test_df.shape
```
