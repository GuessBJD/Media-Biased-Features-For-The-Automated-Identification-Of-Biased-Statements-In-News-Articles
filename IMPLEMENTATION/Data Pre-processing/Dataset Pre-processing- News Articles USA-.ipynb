```python
import pandas as pd
import numpy as np
import re

from sklearn.model_selection import train_test_split
```

```python
df_load = pd.read_csv("Datasets/news_headlines_with_scrap_news_summary.csv")
```

```python
df = df_load.copy()
```

```python
df
```

```python
for i in range(10500):
  text_raw = df.iloc[i]['news_summary']
  text_processed = re.sub("\n"," ", text_raw)
  text_processed = re.sub("https?:\/\/.*[\r\n]*", "", text_processed)
  text_processed = re.sub("@\S+", "", text_processed)
  text_processed = re.sub("\[.*?\]", "", text_processed)
  text_processed = re.sub("\s{2,}", " ", text_processed)
  df.at[i,'news_summary'] = text_processed
```

```python
test = df.loc[7000, 'news_summary']
```

```python
print(test)
```

```python
df = df.drop(['title', 'url'], axis=1)
df.rename(columns={'news_summary': 'text'}, inplace=True)
```

```python
df.insert(0, 'text', df.pop('text'))
df
```

```python
df['slant'].value_counts()
```

```python
train_df, test_df = train_test_split(df, test_size=0.30, random_state=42)
train_df.shape, test_df.shape
```

```python
train_df.to_csv('Datasets/Splits/Train/news_article_summary_usa_label_hyperpartisan_train.csv', index = False)
```
